```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast
```

```python
#read all files in
```

```python
with open('sc2ngQ_v_sc02ngQ.txt') as f:
    sc2ngQ_v_sc02ngQ = f.read().splitlines()

with open('sc2ngQ_v_sc2ngQ.txt') as f:
    sc2ngQ_v_sc2ngQ = f.read().splitlines()
    
with open('sc02ngQ_v_sc02ngQ.txt') as f:
    sc02ngQ_v_sc02ngQ = f.read().splitlines()
    
with open('bulkQ_v_bulkQ.txt') as f:
    bulkQ_v_bulkQ = f.read().splitlines()

with open('sc2ngQ_v_sc2ngM.txt') as f:
    sc2ngQ_v_sc2ngM = f.read().splitlines()
    
with open('sc2ngQ_v_Bulk.txt') as f:
    sc2ngQ_v_Bulk = f.read().splitlines()
    
    
sc2ngQ_v_sc02ngQ = [float(i) for i in sc2ngQ_v_sc02ngQ]
sc2ngQ_v_sc2ngQ = [float(i) for i in sc2ngQ_v_sc2ngQ]
sc02ngQ_v_sc02ngQ = [float(i) for i in sc02ngQ_v_sc02ngQ]
bulkQ_v_bulkQ = [float(i) for i in bulkQ_v_bulkQ]
sc2ngQ_v_sc2ngM = [float(i) for i in sc2ngQ_v_sc2ngM]
sc2ngQ_v_Bulk = [float(i) for i in sc2ngQ_v_Bulk]

all_df = pd.DataFrame({'2ngQ v. 2ngQ': pd.Series(sc2ngQ_v_sc2ngQ), 
                   '02ngQ v. 02ngQ': pd.Series(sc02ngQ_v_sc02ngQ),
                       '2ngQ v. 02ngQ': pd.Series(sc2ngQ_v_sc02ngQ),
                       'BulkQ v. BulkQ': pd.Series(bulkQ_v_bulkQ),
                       '2ngQ v. BulkQ': pd.Series(sc2ngQ_v_Bulk),
                   '2ngQ v. 2ngM': pd.Series(sc2ngQ_v_sc2ngM)
                   
                  })
```

```python
plt.rcParams["figure.dpi"] = 600
fig = all_df.plot.kde(figsize=(10, 5)).get_figure()
plt.xlabel('Cosine Score')
fig.savefig('spec_comp.png')
```
