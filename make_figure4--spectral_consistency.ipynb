```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast
```

```python
with open('data/scores_for_figure4/sc2ngQ_v_sc2ngQ.txt') as f:
    sc2ngQ_v_sc2ngQ = f.read().splitlines()

with open('data/scores_for_figure4/sc02ngQ_v_sc02ngQ.txt') as f:
    sc02ngQ_v_sc02ngQ = f.read().splitlines()
    
with open('data/scores_for_figure4/scQ_v_scQ.txt') as f:
    scQ_v_scQ = f.read().splitlines()
    
with open('data/scores_for_figure4/sc2ngQ_v_Bulk.txt') as f:
    sc2ngQ_v_Bulk = f.read().splitlines()

with open('data/scores_for_figure4/sc02ngQ_v_Bulk.txt') as f:
    sc02ngQ_v_Bulk = f.read().splitlines()

with open('data/scores_for_figure4/scQ_v_Bulk.txt') as f:
    scQ_v_Bulk = f.read().splitlines()
    
sc2ngQ_v_sc2ngQ = [float(i) for i in sc2ngQ_v_sc2ngQ]
sc02ngQ_v_sc02ngQ = [float(i) for i in sc02ngQ_v_sc02ngQ]
scQ_v_scQ = [float(i) for i in scQ_v_scQ]
sc2ngQ_v_Bulk = [float(i) for i in sc2ngQ_v_Bulk]
sc02ngQ_v_Bulk = [float(i) for i in sc02ngQ_v_Bulk]
scQ_v_Bulk = [float(i) for i in scQ_v_Bulk]

all_df = pd.DataFrame({'2ngQ v. 2ngQ': pd.Series(sc2ngQ_v_sc2ngQ), 
                   '02ngQ v. 02ngQ': pd.Series(sc02ngQ_v_sc02ngQ),
                       'scQ v. scQ': pd.Series(scQ_v_scQ),
                       '2ngQ v. BulkQ': pd.Series(sc2ngQ_v_Bulk),
                       '0.2ngQ v. BulkQ': pd.Series(sc02ngQ_v_Bulk),
                   'scQ v. BulkQ': pd.Series(scQ_v_Bulk)
                   
                  })
```

```python
all_df.head()
```

```python
color_dict = {'2ngQ v. 2ngQ': '#85b3e6', '02ngQ v. 02ngQ': '#873e23', 'scQ v. scQ':'#21130d',
             '2ngQ v. BulkQ': '#85b3e6', '0.2ngQ v. BulkQ': '#873e23', 'scQ v. BulkQ':'#21130d'}

plt.rcParams["figure.dpi"] = 600
ax = all_df.plot.kde(color=[color_dict.get(x, '#333333') for x in all_df.columns])
ax.lines[0].set_linestyle("-")
ax.lines[1].set_linestyle("-")
ax.lines[2].set_linestyle("-")
ax.lines[3].set_linestyle("--")
ax.lines[4].set_linestyle("--")
ax.lines[5].set_linestyle("--")

leg = ax.legend()
leg_lines = leg.get_lines()

# ax.set_xlim(-1, 17)
ax.set_title("")
plt.xlim(-.25,1.1)
plt.xlabel('Cosine Score')
fig1 = plt.gcf()
plt.show()
```
