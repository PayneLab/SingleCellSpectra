```python
def combine_peptide_info(df, quality):
    # combine_peptide_and_charge
    df['peptide_charge'] = df['peptide'] + "_" + df['Precursor Charge'].astype(str)

    #combine the scans on each peptide
    df_scan = df.groupby(['peptide_charge'])['scan'].apply(lambda x: ','.join(x.astype(str))).reset_index()
    
   
    #combine the probability of each peptide
    df_prob = df.groupby(['peptide_charge'])['QValue'].apply(lambda x: ','.join(x.astype(str))).reset_index()
  
    
    #merge the scan info with the retention time info
    df_new = pd.merge(df_scan,df_prob, on='peptide_charge')

    
    #set peptide as the index
    df_new = df_new.set_index("peptide_charge")
 
    return df_new
```

```python
def get_combinations_quality_sametype(row):
    file_names = row.index.levels[0].values
    scans = []
    for file in file_names:
        scan_prob = row[file]['scan:prob']
        scan = row[file]['scan']

        if isinstance(scan, str): #if the scan isn't NAN
            #get a list of scans and a list of prob
            scan_probs = scan_prob.split(',') 
            str_suff  = ':' + file
            scan_probs = [sub + str_suff for sub in scan_probs]
            scans.append(scan_probs)

    scans = [item for sublist in scans for item in sublist] #convert list of lists to list


    good_combinations = []
    problist = []
    for el in scans:
        problist.append(float(el.split(':')[1]))

       
    if any(prob <= 0.01 for prob in problist): 
        combos = list(itertools.combinations(scans, 2))
        for tuple_combo in combos:
            prob1 = float(tuple_combo[0].split(':')[1])
            prob2 = float(tuple_combo[1].split(':')[1])

            if (prob1 <= 0.01) & (prob2 <= 0.01):
                good_combinations.append(tuple_combo)


    return good_combinations

def get_combinations_quality_mixedtypes(row):
    file_names = row.index.levels[0].values
    scans = []
    for file in file_names:
        scan_prob = row[file]['scan:prob']
        scan = row[file]['scan']

        if isinstance(scan, str): #if the scan isn't NAN
            #get a list of scans and a list of RT and a list of prob
            scan_probs = scan_prob.split(',') 
            str_suff  = ':' + file
            scan_probs = [sub + str_suff for sub in scan_probs]
            scans.append(scan_probs)


    #we only want to compare scans that have the same retention time
    scans = [item for sublist in scans for item in sublist]

    
    good_combinations = []
    problist = []
    file_type = []
    for el in scans:
        problist.append(float(el.split(':')[1]))
        file_type.append(el.split(':')[2].split('_')[1])


    if any(prob <= 0.01 for prob in problist): 
        if len(set(file_type)) >= 2: #we don't want to compare like file types
            combos = list(itertools.combinations(scans, 2))
            for tuple_combo in combos:
                prob1 = float(tuple_combo[0].split(':')[1])
                prob2 = float(tuple_combo[1].split(':')[1])

                file_name1 = tuple_combo[0].split(':')[2].split('_')[1]
                file_name2 = tuple_combo[1].split(':')[2].split('_')[1]

                if file_name1 != file_name2: #we don't want to compare like file types
                    if (prob1 <= 0.01) & (prob2 <= 0.01):
                        good_combinations.append(tuple_combo)


        else: 
            pass

                
    return good_combinations

```

```python
def get_first_scan(row):
    return row['value'][0]

def get_second_scan(row):
    return row['value'][1]


def separate_into_scan_and_file(df):
    
#     df = df['comparisons'].apply(pd.Series).reset_index().melt(id_vars='index').dropna()[['index', 'value']].set_index('index')
    df = df['comparisons'].apply(pd.Series).reset_index().melt(id_vars='peptide_charge').dropna()[['peptide_charge', 'value']].set_index('peptide_charge')


    df['scan1'] = df.apply(get_first_scan, axis=1)
    df['scan2'] = df.apply(get_second_scan, axis=1)
    df = df.sort_index()

    df[['Scan1','Prob1','File1']] = df['scan1'].str.split(":", expand=True)
    df[['Scan2','Prob2','File2']] = df['scan2'].str.split(":", expand=True)
    df['Prob1'] = pd.to_numeric(df['Prob1'])
    df['Prob2'] = pd.to_numeric(df['Prob2'])
    df = df.drop(columns = ['value', "scan1", 'scan2'])
    
    return df
```
