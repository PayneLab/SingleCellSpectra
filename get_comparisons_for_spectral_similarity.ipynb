```python
def combine_peptide_info(df, quality):
    #combine the scans on each peptide
    df_scan = df.groupby(['peptide'])['scan'].apply(lambda x: ','.join(x.astype(str))).reset_index()
    
    #combine the retention time on each peptide
    df_rt = df.groupby(['peptide'])['minute'].apply(lambda x: ','.join(x.astype(str))).reset_index()

    #combine the retention time on each peptide
    df_prob = df.groupby(['peptide'])['probability'].apply(lambda x: ','.join(x.astype(str))).reset_index()
  
    
    #merge the scan info with the retention time info
    df_new = pd.merge(df_scan,df_rt, on='peptide')
    df_new = pd.merge(df_new,df_prob, on='peptide')
 
    #set peptide as the index
    df_new = df_new.set_index("peptide")
 
    return df_new
```

```python
def combine_scan_prob(row):
    scans = row['scan'].split(',')
    prob = row['probability'].split(',')
    prob_scans = ""
    for i in range(len(scans)):
        both = scans[i] + ":" + prob[i]
        prob_scans = prob_scans + both
        if i < (len(scans)-1):
            prob_scans = prob_scans + ","
    return prob_scans
```

```python
def get_first_scan(row):
    return row['value'][0]

def get_second_scan(row):
    return row['value'][1]


def separate_into_scan_and_file(df):
    
    df = df['comparisons'].apply(pd.Series).reset_index().melt(id_vars='index').dropna()[['index', 'value']].set_index('index')
    
    df['scan1'] = df.apply(get_first_scan, axis=1)
    df['scan2'] = df.apply(get_second_scan, axis=1)
    df = df.sort_index()

    df[['Scan1','Prob1','File1']] = df['scan1'].str.split(":", expand=True)
    df[['Scan2','Prob2','File2']] = df['scan2'].str.split(":", expand=True)
    df['Prob1'] = pd.to_numeric(df['Prob1'])
    df['Prob2'] = pd.to_numeric(df['Prob2'])
    df = df.drop(columns = ['value', "scan1", 'scan2'])
    
    return df
```
